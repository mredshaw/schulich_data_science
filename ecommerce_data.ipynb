```python
grouped_brand_event_counts = df.groupby('brand')['event_type'].count()
```

```python
grouped_brand_event_counts_top_10 = grouped_brand_event_counts.sort_values(ascending = False).head(10)
```

```python
purchase_counts_per_brand = df[df['event_type'] == 'purchase'].groupby('brand').size()

print(purchase_counts_per_brand)
```

```python
df[df['brand'] == 'constancy']
```

```python
brand_order = ['samsung', 'apple', 'xiaomi', 'huawei', 'cordiant', 'lucente', 'lg', 'bosch', 'oppo', 'sony']

purchase_counts_per_brand_top_10 = purchase_counts_per_brand.loc[brand_order]

print(purchase_counts_per_brand_top_10)
```

```python
brand_sale_conversion_rate = purchase_counts_per_brand_top_10 / grouped_brand_event_counts_top_10
print(brand_sale_conversion_rate)
```

```python
brand_sale_conversion_rate.sort_values(ascending=False)
```

```python
top_products_per_brand = df[df['event_type'] == 'purchase'].groupby(['brand', 'product_id']).size().reset_index(name='purchase_count')
max_purchase_per_brand = top_products_per_brand.groupby('brand')['purchase_count'].max()
most_purchased_products = top_products_per_brand[top_products_per_brand['purchase_count'].isin(max_purchase_per_brand)]

print(most_purchased_products)
```

```python
most_purchased_products[most_purchased_products['brand'] == 'apple'].sort_values(by = 'count', ascending= False)
```
