```python
import pandas as pd
import numpy as np

# Creating a sample DataFrame with missing values
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Dave'],
        'Age': [25, np.nan, 30, 35],
        'Salary': [50000, 60000, np.nan, 70000]}
df = pd.DataFrame(data)

# Listwise deletion: Removing rows with any missing values
df_listwise_deletion = df.dropna()

# Pairwise deletion: Keeping rows with at least one non-missing value
df_pairwise_deletion = df.dropna(how='all')

# Imputation: Filling missing values with a constant value
df_imputation = df.fillna(0)

# Interpolation: Filling missing values using linear interpolation
df_interpolation = df.interpolate()

print("Original DataFrame:")
print(df)
print("\nListwise Deletion:")
print(df_listwise_deletion)
print("\nPairwise Deletion:")
print(df_pairwise_deletion)
print("\nImputation:")
print(df_imputation)
print("\nInterpolation:")
print(df_interpolation)
```
